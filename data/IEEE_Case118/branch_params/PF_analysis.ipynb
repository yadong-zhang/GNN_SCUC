```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PF_dir = '../outputs/PF/'
csv_files = [f for f in os.listdir(PF_dir) if f.endswith('.csv')]

PF = []
for file in csv_files:
    pf = pd.read_csv(PF_dir+file, header=None, index_col=None)
    PF.append(pf.abs())

PF = pd.concat(PF, axis=1, ignore_index=True)
```

```python
PF_max = PF.max(axis=1)
```

```python
plt.figure(figsize=(5, 3))
plt.plot(PF_max, label='max')
plt.legend()
plt.tight_layout()
plt.show()
plt.close('all')
```

```python
(PF_max>=1000).astype(int).to_csv('PF_max_category1.csv', index=False, header=False)
((PF_max<1000) & (PF_max>=500)).astype(int).to_csv('PF_max_category2.csv', index=False, header=False)
(PF_max<500).astype(int).to_csv('PF_max_category3.csv', index=False, header=False)
```
